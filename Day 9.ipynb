```python
import numpy as np

def percentage_lower_upper_sec(ls, lower, upper):
    return len([i for i in ls if lower<=i and i<=upper])/len(ls)

a = np.random.normal(60,10, 100000)
print(percentage_lower_upper_sec(a, 50, 70))
```

```python
#Our solution for Normal Distribution - Z
def a_scaledd(ls):
    mean = np.mean(ls)
    std = np.std(ls)
    list=[]
    for num in ls:
        list.append((mean - num)/std)
    print(list)
    
a_scaledd(a)
```

```python
#Milad's solution for Z distribution - is a normal distribution with 0 mean and 1 as its std
import seaborn as sns

a_scaled = [(i-np.mean(a))/np.std(a) for i in a]
# print(a_scaled)
sns.distplot(a_scaled, hist=True, kde=True, bins=20)
```

```python
def fx(x, mu, sigma):
    return (1/(np.sqrt(2*np.pi)*sigma))*np.exp((-(x-mu)**2)/2*sigma**2)


def gaussian_f(x, mu, sigma):
    f = 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(x - mu)**2/2/(sigma**2))
    return f
```

```python
from sympy import symbols, diff
# initialize x and y to be symbols to use in a function
x, y = symbols('x y', real=True)
f = (x**2)/y
# Find the partial derivatives of x and y
fx = diff(f, x, evaluate=True)
fy = diff(f, y, evaluate=True)
print(fx)
print(fy)
# print(f.evalf(subs={x: 2, y: 1}))

print(fx.evalf(subs={x: 2, y: 1}))
print(fy.evalf(subs={x: 2, y: 1}))
```

```python
#Analytical Derivative calculation //2h3m
from sympy import symbols, diff
x = symbols('x', real=True)
f = (x**2)
fx = diff(f, x, evaluate=True)
print(fx)
print(fx.evalf(subs={x: 2}))
print(fx.evalf(subs={x: -3}))
```
